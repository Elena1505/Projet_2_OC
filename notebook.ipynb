Je constate que les données s'étandent sur 130 ans, je vais séléctionner une plus courte période pour travailler ( de 2004 à 2014) et supprimer les autres. Au début j'ai pris de 2010 à 2020, mais il y avait beaucoup trop de valeurs manquantes sur les trois dernières années (j'ai remarqué plus tard que les dernières données ont été enregistrées en 2016).
Je supprime également les colonnes qui ne m'interessent pas. Je visualise les 5 premières lignes pour voir si ça me convient.

```python
data.drop(data.loc[:, "1970":"2003"].columns, axis = 1, inplace = True)
data.drop(data.loc[:, "2015":"2100"].columns, axis = 1, inplace = True)
data = data.drop(columns = "Country Name")
data = data.drop(columns = "Indicator Code")
data = data.drop(columns = "Unnamed: 69")
data.head()
```

J'ai sélectionné 5 variables en tout, je vais faire apparaître seulement ces 5 variables dans mon DataFrame. J'affiche encore les 5 premières lignes pour voir où j'en suis.

```python
data = data.loc[(data["Indicator Name"] == "GDP per capita (current US$)")
                |(data["Indicator Name"] == "Enrolment in upper secondary education, both sexes (number)")
                |(data["Indicator Name"] == "Enrolment in tertiary education, all programmes, both sexes (number)")
                |(data["Indicator Name"] == "Population, ages 15-24, total")
                |(data["Indicator Name"] == "Internet users (per 100 people)"), :]
data.head()
```

J'affiche l'indicateur "Interet users" pour l'Inde de 2004 à 2014. Je reviendrais dessus à la fin.

```python
inde = data.loc[(data["Indicator Name"] == "Internet users (per 100 people)")
              &(data["Country Code"] == "IND"), :]
inde
```

Je décide de calculer la moyenne des variables pour toutes les années afin de travailler avec une unique valeur pour la suite des calculs. Pour cela, j'agrège mes lignes, j'applique ma fonction et je visualise mes 5 premières lignes.

```python
data = data.groupby(["Indicator Name","Country Code"]).sum()
data = data.mean(axis=1)
data.head()
```

Je fais un "reset_index" car je vais avoir besoin de mes variables "Country Code" et "Indicator Name".

```python
data = data.reset_index()
data.head()
```

Je renomme ma nouvelle colonne ainsi que celle des pays à l'aide de la fonction "rename" pour plus de facilité.

```python
data = data.rename(columns = {0:"mean_years"})
data = data.rename(columns = {"Country Code":"Country"})
data.head()
```

Je génère un DataFrame avec la variable "population" afin de faire un premier tri parmis mes pays. 
Je commence par trier mon DataFrame dans l'ordre décroissant à l'aide de la fonction "sort_values".

```python
population = data.loc[(data["Indicator Name"] == "Population, ages 15-24, total"), :]
population = population.sort_values("mean_years", ascending = False)
population.head()
```

Je rajoute une colonne à mon tabelau avec les pourcentage cumulés croissants. 

```python
population['cum_percent'] = 100*(population.mean_years.cumsum() / population.mean_years.sum())
population.head()
```

Je vais garder le top 80% des pays dont la population des 15 à 24 ans est la plus importante.

```python
population = population.loc[(population["cum_percent"])<=80, :]
population
```

Je génère une liste avec les pays qui vérifient mon critère à l'aide de la fonction "tolist".

```python
lst = population["Country"].tolist()
lst
```

Je filtre mon DataFrame principal avec les pays intéressants.
Je vérifie : j'ai bien 160 lignes (32 pays * 5 variables).

```python
data = data[data.Country.isin(lst)]
data
```

Je vais maintenant me servir du fichier "EdStatsCountry.csv" (data2 dans mon notebook) pour étudier les niveaux de revenus et pour savoir à quelle région du monde appartient quel pays. 
Je décide de garder 4 niveaux de revenus sur 5 pour faire un deuxième filtre. En effet, je considère que les pays qui ont un revenu faible ne peuvent pas être ciblés par l'entreprise.

```python
data2 = data2.loc[(data2["Income Group"] == "High income: OECD")
                 |(data2["Income Group"] == "High income: nonOECD")
                 |(data2["Income Group"] == "Upper middle income")
                 |(data2["Income Group"] == "Lower middle income"), 
                  ["Country Code", "Region", "Income Group"]]

data2.head()
```

Je renomme ici aussi ma colonne "Country Code" pour avoir une clé commune pour fusionner mes deux DataFrame.

```python
data2 = data2.rename(columns = {"Country Code":"Country"})
data2.head()
```

Je fusionne mes deux dataframe à l'aide de la fonction "merge" afin de les étudier ensemble.

```python
data = pd.merge(data, data2, on = "Country")
data.head()
```

Maintenant que j'ai séléctionné les pays qui m'intéressent, je vais générer les statistiques descriptives classiques à l'aide de la fonction "describe" pour chaque région du monde.
Je vais créer un DataFrame par région, soit 7 et appliquer la fonction à l'aide d'une boucle "for".

```python
data_latin_america_and_caribbean = data.loc[(data["Region"] == "Latin America & Caribbean"), :]
data_north_africa = data.loc[(data["Region"] == "North America"), :]
data_middle_est_and_north_africa = data.loc[(data["Region"] == "Middle East & North Africa"), :]
data_south_asia = data.loc[(data["Region"] == "South Asia"), :]
data_sub_saharan_africa = data.loc[(data["Region"] == "Sub-Saharan Africa"), :]
data_east_asia_and_pacific = data.loc[(data["Region"] == "East Asia & Pacific"), :]
data_europe_and_central_asia = data.loc[(data["Region"] == "Europe & Central Asia"), :]

regions = [data_latin_america_and_caribbean, data_north_africa, data_middle_est_and_north_africa, data_south_asia, data_sub_saharan_africa, data_east_asia_and_pacific, data_europe_and_central_asia]

for i in regions: 
    print(i.describe(), "\n")
```

En comparant les moyennes, médianes et écarts-types des différentes régions du monde, on remarque que celles qui se démarquent sont l'Amérique latine et les Caraîbes, et, l'Europe et l'Asie centrale.
